3. Постройте изображение с настройками по умолчанию в Matplotlib:

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sinplot()
plt.show()
```

4. Измените тот же график на значения по умолчанию в Seaborn, используя функцию set():

```python
def sinplot(flip = 1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sb.set()
sinplot()
plt.show()
```

разница в визуальной части оформления 

5. Применим стиль "whitegrid"

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 5):
        plt.plot(x, np.sin(x + i*.5)*(7 - i)*flip)
sb.set_style("whitegrid")
sinplot()
plt.show()
```

разница между 4 и 5 примером:
