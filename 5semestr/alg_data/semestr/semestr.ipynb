6. (8 баллов) Проведите вычислительные эксперименты и нарисуйте графики,
показывающие зависимость времени выполнения от размера входных данных n.
Возьмите n равное: 100, 500, 1000, 2000, 4000, 6000, 8000, 10000.
Реализуйте функции для генерации следующих вариантов входных данных: 

```python
ranges = [100, 500, 1000, 2000, 4000, 6000, 8000, 10000]
plt_x_all, plt_t_all = [], []
```

a. список случайных чисел; 

```python
r_list = lambda r: [random.random() for _ in range(r)]

plt_x, plt_t = [], []
timer = Timer(f"merge_sort(a)", globals=globals())

for r in ranges:
    print(r)
    merge_sort(some_list)
    a = r_list(r)
    plt_x.append(r)
    t = timer.timeit(number=100)
    plt_t.append(t)
    plt_t_all.append(t)

plt.plot(plt_x, plt_t, marker='.', label="random")
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```

b. почти отсортированный список (один элемент в середине списка находится
не на своем месте); 

```python
def almost(r):
    a = list(range(r))
    x, y = r//2, r//4
    a[x], a[x + y] = a[x + y], a[x]
    return a

plt_x, plt_t = [], []
timer = Timer(f"merge_sort(a)", globals=globals())

for r in ranges:
    print(r)
    merge_sort(some_list)
    a = almost(r)
    plt_x.append(r)
    t = timer.timeit(number=100)
    plt_t.append(t)
    plt_t_all.append(t)
    

plt.plot(plt_x, plt_t, marker='.', label="almost")
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```

c. почти отсортированный в обратную сторону список (один элемент в середине
списка находится не на своем месте);

```python
def almost_reverse(r):
    a = list(range(r))
    a.reverse()
    x, y = r//2, r//4
    a[x], a[x + y] = a[x + y], a[x]
    return a

plt_x, plt_t = [], []
timer = Timer(f"merge_sort(a)", globals=globals())

for r in ranges:
    print(r)
    merge_sort(some_list)
    a = almost_reverse(r)
    plt_x.append(r)
    t = timer.timeit(number=100)
    plt_t.append(t)
    plt_t_all.append(t)

plt.plot(plt_x, plt_t, marker='.', label="almost_reverse")
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```

e. первые 95% списка упорядочены, а последние 5% содержат случайные
значения;


```python
def list_95(r):
    a = list(range(r))
    idx = int(r * 0.95)
    a[idx:] = [random.random() for _ in range(r - idx)]
    return a

plt_x, plt_t = [], []
timer = Timer(f"merge_sort(a)", globals=globals())

for r in ranges:
    print(r)
    merge_sort(some_list)
    a = list_95(r)
    plt_x.append(r)
    t = timer.timeit(number=100)
    plt_t.append(t)
    plt_t_all.append(t)

plt.plot(plt_x, plt_t, marker='.', label="list_95")
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```

f. все элементы находятся в пределах 10 позиций от окончательного места в
массиве.

```python
def var10(r):
    a = list(range(r))
    chunks = [10 * x for x in range(r//10)]
    for chunk in chunks:
        aa = a[chunk:chunk+10]
        random.shuffle(aa)
        a[chunk:chunk+10] = aa
    return a

plt_x, plt_t = [], []
timer = Timer(f"merge_sort(a)", globals=globals())

for r in ranges:
    print(r)
    merge_sort(some_list)
    a = var10(r)
    plt_x.append(r)
    t = timer.timeit(number=100)
    plt_t.append(t)
    plt_t_all.append(t)

plt.plot(plt_x, plt_t, marker='.', label="var10")
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```

```python
import numpy as np

plt_x_all = plt_x
plt_t_all = np.array(plt_t_all)
plt_t_all = plt_t_all.reshape(6,8)
title = ["random","almost","almost_reverse","zero_one","list_95","var10"]
for i, j in enumerate(plt_t_all):
    plt.plot(plt_x_all, j, marker='.', label=title[i])
plt.xlabel("elements")
plt.ylabel("time")
plt.legend()
plt.show()
```
