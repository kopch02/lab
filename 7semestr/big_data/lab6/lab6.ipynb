> ## EX-4-2
>
> Напишите программу, которая считывает целое неотрицательное число n и выводит его факториал. Факториал числа n обозначается n! и равен произведению всех натуральных чисел от 1 до n.

```python
n = int(input("->"))
if n > 0:
    s = 1
    for i in range(1,n+1):
        s *= i
    print(s)
```

> ## EX-4-4
>
> Выведите пирамиду из символов «*» заданной высоты. Пирамиду, не прямоугольный треугольник :)

```python
h = int(input("введите высоту:"))
n = h
g = 1
for i in range(h):
    n -= 1
    print(" " * n, "*" * g, " " * n)
    g += 2
```

> ## EX-4-5
>
> Напишите программу построения прямоугольника размерами h х w, состоящего из символов symb. h, w  и symb - вводит пользователь на отдельных строках.

```python
w = int(input("w = "))
h = int(input("h = "))
symb = input("symb = ")
print(symb * w)
if h > 1:
    if h > 2:
        for i in range(h - 2):
            print(symb + " " * (w - 2) + symb)
    print(symb * w)
```
