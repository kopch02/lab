```python
merged_df = df_vgs.merge(df_metacritic[['name', 'rating']], how='left', left_on='Name', right_on='name')
merged_df.rename(columns={'rating': 'metacritic_rating'}, inplace=True)
merged_df.drop('name', axis=1, inplace=True)
merged_df.head()
```

3. Вывести список игр, рейтинг которых равен "M" и год издания не ранее 2012 года;

```python
dop_df = merged_df[(merged_df['metacritic_rating']=='M') & (merged_df["Year"]>=2012)]
dop_df
```

5. Вывести жанры игр с количеством игр, которые принадлежат соответствующему жанру в виде "<жанр> - <количество игр>" для всех жанров, которые содержат в своем названии не менее 3 различных гласных.

```python
def find_gener(word):
    vowels = set("aeiouAEIOU")
    count = 0
    for char in word:
        if char in vowels:
            count += 1
    return count

filtered_genres = df_vgs[df_vgs['Genre'].apply(lambda x: find_gener(x) >= 3)]#применяем функцию к каждому жанру
genre_counts_filtered = filtered_genres['Genre'].value_counts()

for genre, count in genre_counts_filtered.items():
    print(f"<{genre}> - <{count}>")
```
