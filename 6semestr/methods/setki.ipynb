```python
import numpy as np


def F(x):
    return 2 * x ** 2

def P(x):
    return -0.5

def Q(x):
    return 3

a = 1
b = 1.3
n = 10
k1 = 1
k2 = 2
l1 = 1
l2 = 0
R1 = 0.6
R2 = 1

h = (b - a) / n

xi = []

for i in range(n + 1):
    xi.append(a + h * i)

Ai = []
Bi = []
Ci = []

for i in range(n + 1):
    Ai.append(1 / (h * h) - P(xi[i]) / (2 * h))
    Bi.append(2 / (h * h) - Q(xi[i]))
    Ci.append(1 / (h * h) + P(xi[i]) / (2 * h))

v1 = -k2 / (k1 * h - k2)
w1 = (h * R1) / (k1 * h - k2)

v2 = l2 / (l1 * h + l2)
w2 = (h * R2) / (l1 * h + l2)

alph = [v1]
bet = [w1]

for i in range(n):
    alph.append(-Ci[i + 1] / (Ai[i + 1] * alph[i] - Bi[i + 1]))
    bet.append((F(xi[i+1]) - Ai[i + 1] * bet[i]) / (Ai[i + 1] * alph[i] - Bi[i + 1]))

y = [(v2 * bet[n] + w2) / (1 - v2 * alph[n])]

for i in range(n - 1, -1, -1):
    y.append(alph[i] * y[n - (i + 1)] + bet[i])


y.reverse()


print('Краевая задача. Метод Сеток''')
print('y`` + 0,6xy` - 2y = 1')
print('Краевые условия:')
print('         y(1,5) = 0,6')
print('         2y(1,8) - 0,8y`(1,8) = 3')
print(f"{a = }")
print(f"{b = }")
print(f"{h = }")
print(f"{n = }")
print(f"{k1 = }")
print(f"{k2 = }")
print(f"{l1 = }")
print(f"{l2 = }")
print(f"{R1 = }")
print(f"{R2 = }")
print(f"{Ai = }")
print(f"{Bi = }")
print(f"{Ci = }")
print('Fi = 2*xi**2')
print(f"{v1 = }")
print(f"{w1 = }")
print(f"{v2 = }")
print(f"{w2 = }")

print('--------------------------------------------------------------------------------------------------------------')
print(' x\t\t\tyi\t\t\t\t\talpha\t\t\t\t\tbeta')
for i in range(n + 1):
    print(f"{round(xi[i],2)}\t\t\t{y[i]}\t\t\t{alph[i]}\t\t\t{bet[i]}")

kray1 = k1 * y[0] + k2 * ((y[1] - y[0])/h)
kray2 = l1 * y[n] + l2 * ((y[n] - y[n - 1])/h)

print(f"{kray1 = }")
print(f"{kray2 = }")
```
